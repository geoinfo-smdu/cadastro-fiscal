```python
import geopandas as gpd
import pandas as pd
```

```python
from pathlib import Path

gdf = gpd.GeoDataFrame({'SQLC-lote': []})

for path in Path('lotes').rglob('*.zip'):

    gdf_temp = gpd.GeoDataFrame.from_file(f'zip://{path}!{path.name[:-4]}')
    
    print(path.name[:-4], len(gdf_temp))
    
    gdf_temp.loc[(gdf_temp['lo_condomi'] == '00') & \
                 (gdf_temp['lo_tp_lote'] == 'F'), 'lote'] = \
    gdf_temp.loc[(gdf_temp['lo_condomi'] == '00') & \
                 (gdf_temp['lo_tp_lote'] == 'F'), 'lo_lote']
    
    gdf_temp.loc[(gdf_temp['lo_condomi'] != '00') & \
                 (gdf_temp['lo_tp_lote'] == 'F'), 'lote'] = \
    '0000'
    
    gdf_temp['SQLC-lote'] = gdf_temp['lo_setor'] + \
                            gdf_temp['lo_quadra'] + \
                            gdf_temp['lote'] + \
                            gdf_temp['lo_condomi']
    
    gdf_temp.geometry = gdf_temp.geometry.representative_point()
    pd_temp = pd.concat([gdf, gdf_temp[['SQLC-lote', 'geometry']]], ignore_index=True)

    gdf = gpd.GeoDataFrame(pd_temp, geometry=pd_temp.geometry)
```

```python
gdf.crs = 'EPSG:31983'
```

```python
gdf = gdf.to_crs(epsg=4326)
```

```python
gdf['x'] = gdf.geometry.x
```

```python
gdf['y'] = gdf.geometry.y
```

```python
gdf[gdf['SQLC-lote'].notna()].drop(['x', 'y'], axis=1).to_file('resultados/lotes-por-pontos.gpkg', driver='GPKG')
```

```python
pd.DataFrame(gdf.loc[gdf['SQLC-lote'].notna(), \
                     ['SQLC-lote', 'x', 'y']])\
                    .to_csv('resultados/lotes-por-pontos.csv.gz', 
                    compression='gzip', 
                    index=False)
# gdf.loc[gdf['SQLC-lote'].notna(), ['SQLC-lote', 'x', 'y']]
```
